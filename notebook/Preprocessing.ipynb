### Kolom Salary

```python
df["Salary"].value_counts()
```

Menormalisasi salary benar benar tricky karena value nya memiliki rentang, memiliki banyak noise, dan tidak konsisten (ada yang perhour)

```python
df["salary_str"] = df["Salary"].astype(str).str.lower()

# 2. Hilangkan label umum
df["salary_str"] = (
    df["salary_str"]
    .str.replace("rp", "", regex=False)
    .str.replace("idr", "", regex=False)
    .str.replace("/bulan", "", regex=False)
    .str.replace("/ month", "", regex=False)
    .str.replace("per bulan", "", regex=False)
    .str.replace("per month", "", regex=False)
    # .str.replace("/ hour", "", regex=False)
    # .str.replace("per jam", "", regex=False)
    # .str.replace("per hour", "", regex=False)
    .str.replace("\u2005", "", regex=False)
    .str.replace("negotiable", "", regex=False)
)

# 3. Normalisasi whitespace (hapus spasi aneh)
df["salary_str"] = df["salary_str"].str.replace(r"\s+", " ", regex=True)

# 4. Standarisasi pemisah ribuan/desimal
df["salary_str"] = df["salary_str"].str.replace(",", ".", regex=False)
df["salary_str"] = df["salary_str"].str.replace(r"\.(?=\d{3})", "", regex=True)

# 5. Konversi k, m, ribu
df["salary_str"] = df["salary_str"].str.replace(
    r"(\d+(?:\.\d+)?)\s*k", 
    lambda m: str(float(m.group(1)) * 1_000), 
    regex=True
)

df["salary_str"] = df["salary_str"].str.replace(
    r"(\d+(?:\.\d+)?)\s*m", 
    lambda m: str(float(m.group(1)) * 1_000_000), 
    regex=True
)

df["salary_str"] = df["salary_str"].str.replace(
    r"(\d+(?:\.\d+)?)\s*ribu", 
    lambda m: str(float(m.group(1)) * 1_000), 
    regex=True
)

# 6. Trim spasi
df["salary_str"] = df["salary_str"].str.strip()
```

```python
def normalize_range(val):
    if "-" not in val:
        return val
    parts = [p.strip() for p in val.split("-")]
    if len(parts) != 2:
        return val
    
    try:
        left = float(parts[0])
        right = float(parts[1])
    except ValueError:
        return val
    
    # Jika kanan jauh lebih besar (contoh 2000000), dan kiri masih kecil (contoh 1.5)
    if right >= 1_000_000 and left < 100:
        left *= 1_000_000
    elif right >= 1_000 and left < 100:
        left *= 1_000
    
    return f"{left} - {right}"

df["salary_str"] = df["salary_str"].apply(normalize_range)
```

```python
# filter value yang ada tanggal nya
df.loc[df["salary_str"].str.match(r"\d{4}-\d{2}-\d{2}", na=False), "salary_str"] = ""
```

```python
# Mengubah string ke number
import re
def parse_num_str(s):
    s = re.sub(r'[^\d\.,]', '', s)   # keep only digits, dot, comma
    if not s:
        return np.nan

    # heuristik pemisah ribuan/desimal
    if s.count('.') > 1 and s.count(',') == 0:
        # banyak titik → titik = pemisah ribuan
        s = s.replace('.', '')
    elif s.count(',') > 1 and s.count('.') == 0:
        # banyak koma → koma = pemisah ribuan
        s = s.replace(',', '')
    elif s.count('.') >= 1 and s.count(',') >= 1:
        # ada keduanya → lihat posisi terakhir
        if s.rfind('.') > s.rfind(','):
            # titik terakhir setelah koma -> titik dianggap decimal -> hapus koma
            s = s.replace(',', '')
        else:
            # koma terakhir setelah titik -> koma decimal, titik ribuan
            s = s.replace('.', '').replace(',', '.')
    else:
        # hanya satu jenis separator atau tidak ada -> ubah koma jadi titik (decimal)
        s = s.replace(',', '.')

    try:
        return float(s)
    except:
        return np.nan
    
df["numbers"] = df["salary_str"].apply(lambda x: re.findall(r"[\d\.,]+", x))  
df["numbers"] = df["numbers"].apply(lambda lst: [parse_num_str(n) for n in lst] if lst else [])
```

```python
df["numbers"].value_counts()
```

```python
df["salary_min"] = df["numbers"].apply(lambda x: x[0] if len(x)>=1 else np.nan)
df["salary_max"] = df["numbers"].apply(lambda x: x[1] if len(x)>=2 else (x[0] if len(x)==1 else np.nan))
```

```python
# konversi per-hour ke per-month (asumsi 8 jam × 22 hari kerja)
mask_hour = df["salary_str"].str.contains("hour|jam", case=False, na=False)
df.loc[mask_hour, ["salary_min","salary_max"]] = df.loc[mask_hour, ["salary_min","salary_max"]].apply(lambda col: col * 8 * 22)
```

Kolom upload by company memiilki value yang beragam dan mostly dalam bentuk "months, weeks, dan day". Ini perlu dibersihkan dan sebaiknya diubah kedalam bentuk tanggal agar mudah dianalisis nantinya, dan karena ada kolom "access date" itu bisa menjadi acuan menentukan tanggal nya meskipun tidak 100% akurat

```python
from datetime import timedelta

def clean_upload(val, access_date):
    if pd.isna(val):
        return np.nan
    
    # kalau sudah datetime object
    if isinstance(val, pd.Timestamp):
        return val
    
    val = str(val).lower().strip()
    
    val = re.split(r"\(", val)[0].strip()
    
    # special cases
    if val in ["this day", "today"]:
        return access_date
    if "! year ago" in val:
        return access_date - timedelta(days=365)
    
    # regex untuk days, weeks, months
    match = re.search(r"(\d+)\s*(day|days|week|weeks|month|months|moth|moths)", val)
    if match:
        num = int(match.group(1))
        unit = match.group(2)
        if "day" in unit:
            return access_date - timedelta(days=num)
        elif "week" in unit:
            return access_date - timedelta(weeks=num)
        elif "month" in unit or "moth" in unit:  # termasuk typo
            return access_date - timedelta(days=num*30)
    
    # coba parse langsung tanggal (misal 30-Agu-2025)
    parsed = pd.to_datetime(val, errors="coerce", dayfirst=True)
    if not pd.isna(parsed):
        return parsed
    
    return np.nan

# terapkan ke dataframe
df["Upload By Company"] = df.apply(
    lambda row: clean_upload(row["Upload By Company"], row["Access Date"]), axis=1
)
```

### Kolom Enthusiast

membersihkan value yang ada ">" karena ini masih berbentuk object

```python
df["Enthusiast"].value_counts()
```

```python
def clean_enthusiast(val):
    val = str(val).strip()
    if val.startswith(">"):
        return int(val[1:])  # ambil angka setelah ">"
    elif val.isdigit():
        return int(val)
    else:
        return np.nan  # kalau ada yang aneh
        

df["Enthusiast"] = df["Enthusiast"].apply(clean_enthusiast)
```

### Kolom category

kolom ini juga masih tak konsisten ada yang lowercase, menggunakan -, dsb

```python
df["Category"].value_counts()
```

```python
def clean_category(val):
    if pd.isna(val):
        return np.nan
    val = val.lower().strip()
    
    # hapus tanda baca umum
    val = re.sub(r"[-_/]", " ", val)  # ganti -, _, / jadi spasi
    val = re.sub(r"\s+", " ", val)    # rapikan spasi
    
    # mapping manual
    mapping = {
        "full time": "Full-time",
        "part time": "Part-time",
        "contract": "Contract",
        "contractual": "Contract",
        "freelance": "Freelance",
        "intern": "Intern",
    }
    
    return mapping.get(val, val.title())  # fallback: capitalisasi normal

df["Category"] = df["Category"].apply(clean_category)
```

### Kolom Degree

```python
df["Degree"].value_counts()
```

Kolom degree sebenarnya sudah unique semua tapi sebaiknya dirapikan 

```python
def clean_degree(val):
    if pd.isna(val):
        return "Not Specified"
    
    val = val.lower().strip()
    
    if val in ["-", "none"]:
        return "Not Specified"
    if "high school" in val:
        return "High School"
    if "diploma" in val:
        return "Diploma"
    if "bachelor" in val:
        return "Bachelor"
    if "master" in val and "ph" not in val:
        return "Master"
    if "phd" in val or "ph.d" in val:
        return "PhD"
    if "bachelor's or master's" in val:
        return "Bachelor or Master"
    if "master’s or phd" in val:
        return "Master or PhD"
    
    return val.title()

df["Degree"] = df["Degree"].apply(clean_degree)
```
